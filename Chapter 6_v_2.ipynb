```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import matplotlib.pyplot as plt
from sklearn import tree
import graphviz
from dtreeviz.trees import dtreeviz
import os
from sklearn.tree import DecisionTreeRegressor
```

```python
iris = load_iris()
X = iris.data[:,2:]
y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth= 2)
tree_clf.fit(X,y)
```

```python
text_representation = tree.export_text(tree_clf)
print(text_representation)
```

```python
feature_name = list(iris.feature_names[2:])
print(feature_name)
class_name = list(iris.target_names)
print(class_name)
```

```python
fig = plt.figure(figsize=(25,20))
_=tree.plot_tree(tree_clf,fontsize=20,filled= True,rounded= True,feature_names= feature_name,
               class_names= class_name,
              )
fig.savefig("decistion_tree.png")
```

```python
tree_clf.predict_proba([[5,1.5]])
```

```python
tree_clf.predict([[5,1.5]])
```

```python
tree_clf = DecisionTreeClassifier(max_depth= 2,criterion= 'entropy')
tree_clf.fit(X,y)

fig = plt.figure(figsize=(25,20))
_=tree.plot_tree(tree_clf,fontsize=20,filled= True,rounded= True,feature_names= feature_name,
               class_names= class_name,
              )
```

```python
_ = dtreeviz(tree_clf, X, y,
                target_name="target",
                feature_names=feature_name,
                class_names=class_name
            )
_.save('a.svg')
_
```
